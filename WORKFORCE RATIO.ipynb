```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_paygap = pd.read_csv(r"C:\Users\preem\Desktop\PROJECT\DATA ANALYSIS\paygapnew.csv")
df_workratio = pd.read_csv(r"C:\Users\preem\Desktop\PROJECT\DATA ANALYSIS\workratio.csv")
```

```python
df_workforce = pd.merge(df_paygap,df_workratio,on='Occupation',how='outer')
```

```python

lp = sns.barplot(x="Occupation",y="Paygap",data=df_workforce,hue="Ratio",palette="rocket_r",order=df_workforce.sort_values('Paygap').Occupation,dodge=False)
plt.legend(title="Male workforce:Female workforce", loc="best",prop={"size":10})
plt.xticks(rotation=90)
sns.set(rc = {'figure.figsize':(15,8)})
#plt.figure(figsize=(15,10))
save_results_to = "C:/Users/preem/Desktop/PROJECT/"
plt.savefig(save_results_to+"barplot.png", dpi=100)
plt.show()
```
