```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#appending columns of both genders and their professions together through the years(2004-2017)
```

```python
df1 = pd.read_csv(r"C:\Users\preem\Desktop\PROJECT\DATA ANALYSIS\editedm_f.csv")
df1.head()
```

```python
df2 = df1.iloc[:,0:9] #made a new dataframe for females
df2.columns = ["year","gender","managers","professionals","technicians and traders","community service workers","clerical and administrative workers","salesworkers","machinery operators and drivers"]
df2.head()
```

```python
df3 = df1.iloc[:,10:18] #made a new dataframe for males
df3.columns = ["gender","managers","professionals","technicians and traders","community service workers","clerical and administrative workers","salesworkers","machinery operators and drivers"]
dates = pd.Series(pd.date_range("2004",periods = 14,freq ='Y'))
df3.insert(0,'year' ,dates.dt.year,True)
df3.head()
```

```python
df_wages = df2.append(df3,ignore_index=True) #appending both the female and male dataframes
df_wages
```

```python
df_combinedwages = pd.melt(df_wages,id_vars=['gender'],value_vars=["managers","professionals","technicians and traders","community service workers","clerical and administrative workers","salesworkers","machinery operators and drivers"])
```

```python
df_combinedwages
```

```python
df_combinedwages.to_csv(r'C:\Users\preem\Desktop\PROJECT\DATA ANALYSIS\combinedwages.csv', index = False)
```

```python
df_combinedwages.rename(columns = {"variable":"occupation","value":"Average hourly earnings"},inplace = True)
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(y ='occupation',x ='Average hourly earnings', hue ='gender',data = df_combinedwages)
plt.show()
```
